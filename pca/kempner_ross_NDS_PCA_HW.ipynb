```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # For plotting
from sklearn.decomposition import PCA
```

```python
# read in the data
df = pd.read_csv("class6_hw_data.txt", header = None)
df.shape
```

1. Plot the mean pupil response over all trials.



```python

plt.figure()
plt.plot(np.mean(df.to_numpy(),axis = 1))
plt.ylabel("mean pupil response")
plt.xlabel("trials")
plt.title("mean pupil response over trials")
plt.show()
```

```python
df.to_numpy().shape
```

2. Do PCA across trials. Plot the first 3 principal components. (ask yourself if you did the PCA on the right dimension, does it make sense?)

```python
# Perform PCA
pca = PCA()
pcs = pca.fit_transform(df.to_numpy())
explained = pca.explained_variance_ratio_ * 100



# Observation index
x = np.arange(pcs.shape[0])

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, pcs[:, 0], label='PC1')
plt.plot(x, pcs[:, 1], label='PC2')
plt.plot(x, pcs[:, 2], label='PC3')

plt.xlabel('Observation')
plt.ylabel('PC Score')
plt.title('First 3 Principal Components Over Observations')
plt.legend()
plt.show()

```
